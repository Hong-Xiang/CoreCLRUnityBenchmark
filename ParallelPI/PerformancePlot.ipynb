```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Report.csv')
```

```python
df['runtime-method'] = df['runtime'] + "-" + df['method']
```

```python
def plot_data(group):
    g = sns.catplot(data=df[df['group'] == group], kind="bar", x="runtime", y="normalized time(Î¼s)", hue="method")
    g.set(title=f"{group} group")
    return g
```

```python
plot_data("middle")
```

```python
plot_data("small")
```
